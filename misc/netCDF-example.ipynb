### First, create 1/8 degree grid points covering LA county.

These must be identical to the grid points in the netCDF files.

```python
bbox = (240.9375, 33.4375, 243.1875, 34.9375)
step = 0.125

lats = (np.array(np.linspace(bbox[1], bbox[3],
        int(1 + (bbox[3] - bbox[1])/step)), 
        dtype=np.float32))

lons = (np.array(np.linspace(bbox[0], bbox[2],
        int(1 + (bbox[2] - bbox[0])/step)),
        dtype=np.float32) - 360)

latlon = np.vstack(np.dstack(np.meshgrid(lons, lats)))
```

```python
scatter(latlon[:,0], latlon[:,1])
```

### Next, determine which grid points are contained in each climate zone

```python
homedir = os.path.expanduser('~')
climzone = gpd.read_file('%s/Dropbox/CA_Building_Standards_Climate_Zones/CA_Building_Standards_Climate_Zones.shp' % homedir).to_crs(epsg=4326)
```

Below, we are determining the grid points that are contained in each climate zone, and then outputting their indices.

```python
intersection = climzone.set_index('Zone').geometry.apply(lambda x: np.where(vectorized.contains(x, latlon[:,0], latlon[:,1]))[0])
intersection = intersection[intersection.apply(len) > 0]

intersection
```

### Let's rotate the lists in the above series so that we have a single-dimensional series.

```python
# If you want the integer index -- as it would appear in np.meshgrid(lons, lats)
zone_idx = pd.Series(np.concatenate(intersection.values), index=np.repeat(intersection.index.values, intersection.apply(len)))

zone_idx.head(20)
```

### However, what we really want is the x and y indices.

```python
# If you want the x and y indices:

idx_xy = zone_idx.apply(lambda x: np.unravel_index(x, (lons.size, lats.size))).apply(pd.Series).rename(columns={0: 'x', 1: 'y'})

idx_xy.head(20)
```

### Create two functions

The first function concatenates netCDF arrays for a variable of interest (under one projection scenario).

The second function calculates the mean for a given variable over each climate zone.

```python
def concatenate_netcdf_scen(proj_n, var, arglist, time_overlap=None):
    # Assumes the variables are listed [projection, time, lat, lon]
    if time_overlap is not None:
        time_overlap = -time_overlap
    cat = np.concatenate([np.ma.filled(i.variables[var][proj_n][slice(0, time_overlap)][:][:], np.nan) for i in arglist])
    return cat
```

```python
def climate_zone_var(nc, idx_xy, start_date, freq='d', col_labels=['x', 'y']):

    cz_dict = {}
    
    for zone in np.unique(idx_xy.index.values):
        c = idx_xy.loc[zone]
        tmax = np.nanmean(nc[:, c[col_labels[1]].values, c[col_labels[0]].values], axis=1)
        cz_dict.update({ zone : tmax})

    cz_var = pd.DataFrame.from_dict(cz_dict)
    cz_var.index = pd.date_range(start=start_date, periods=len(cz_var), freq='d')
    
    return cz_var
```

```python
# MODEL NO. 116: MPI-ECHAM-LR

cz_tmax = climate_zone_var(concatenate_netcdf_scen(116, 'tasmax', [tmax_2010, tmax_2030, tmax_2050, tmax_2070], time_overlap=31),
                           idx_xy,
                        datetime.date(2010,1,1))

cz_tmax
```

### Let's get the daily temperature profile for a single climate zone

Let's look at climate zone '16'

```python
tmin_h = cz_tmin['16'].resample('h')
tmax_h = cz_tmax['16'].resample('h', loffset=datetime.timedelta(hours=12))
```

```python
daily_T = tmin_h.fillna(tmax_h)

daily_T.head(24)
```

Let's interpolate using a cubic function. To avoid a memory error lets break it into chunks.

```python
def file_chunks(l, n):
    """ Yield successive n-sized chunks from l.
    """
    for i in xrange(0, len(l), n):
        yield l[i:i+n]
```

```python
for i in file_chunks(range(len(daily_T)), 2400):
    if (i[-1] + 1) < len(daily_T):
        i.extend((np.arange(24) + i[-1]).tolist())
    daily_T.iloc[i] = daily_T.iloc[i].interpolate(method='cubic')
```

We now have a good estimate of an hourly temperature profile
