```python
import numpy as np
import pandas as pd
import os
import re
import datetime
import time
from scipy.stats import gaussian_kde
import pysal as ps

basepath = '/home/akagi/Documents/wecc_form_714'

path_d = {
1993: '93WSCC1/WSCC',
1994: '94WSCC1/WSCC1994',
1995: '95WSCC1',
1996: '96WSCC1/WSCC1996',
1997: '97wscc1',
1998: '98WSCC1/WSCC1',
1999: '99WSCC1/WSCC1',
2000: '00WSCC1/WSCC1',
2001: '01WECC/WECC01/wecc01',
2002: 'WECCONE3/WECC One/WECC2002',
2003: 'WECC/WECC/WECC ONE/wecc03',
2004: 'WECC_2004/WECC/WECC One/ferc',
2006: 'form714-database_2006_2013/form714-database/Part 3 Schedule 2 - Planning Area Hourly Demand.csv'
}

util = {
'aps' : 803,
'srp' : 16572,
'ldwp' : 11208
        }
util_2006 = {
'aps' : 116,
'srp' : 244,
'ldwp' : 194

}

df_path_d = {}

def build_paths():
    for y in path_d.keys():
        if y < 2006:
            pathstr = basepath + '/' + path_d[y]
            dirstr = ' '.join(os.listdir(pathstr))
#            print dirstr
            for u in util.keys():
                if not u in df_path_d:
                    df_path_d.update({u : {}})
                srcstr = '%s\d\d.dat' % (u)
 #               print srcstr
                match = re.search(srcstr, dirstr, re.I)
#                print type(match.group())
                rpath = pathstr + '/' + match.group()
                df_path_d[u].update({y : rpath})
        elif y == 2006:
            pathstr = basepath + '/' + path_d[y]
            for u in util.keys():
                if not u in df_path_d:
                    df_path_d.update({u : {}})
                df_path_d[u].update({y : pathstr})

df_d = {}

def build_df():
    for u in df_path_d.keys():
        df = pd.DataFrame()
        for y in df_path_d[u].keys():
            if y < 2006:
#                print df_path_d[u][y]
                f = open(df_path_d[u][y])
                r = f.readlines()
                r = [g.replace('\t', '      ') for g in r if len(g) > 70]
                for i in range(0, len(r)-1, 2):
                    entry = [r[i], r[i+1]]
                    mo = int(r[i][:2])
                    day = int(r[i][2:4])
                    yr = r[i][4:6]
                    if yr[0] == '0':
                        yr = int('20' + yr)
                    else:
                        yr = int('19' + yr)

                    am = [int(j) for j in entry[0][20:].split()]
                    pm = [int(j) for j in entry[1][20:].split()]
                    ampm = am + pm
                    entry_df = pd.DataFrame()
                    dt_ix = pd.date_range(start=datetime.datetime(yr, mo, day, 0), end=datetime.datetime(yr, mo, day, 23), freq='H')
                    entry_df['load'] = ampm
#                    print entry_df
                    entry_df.index = dt_ix
                    df = df.append(entry_df)
#                    print entry_df
            elif y == 2006:
                f = pd.read_csv('%s/%s' % (basepath, path_d[y]))
                f = f.loc[f['respondent_id'] == util_2006[u], [u'plan_date', u'hour01', u'hour02', u'hour03', u'hour04', u'hour05', u'hour06', u'hour07', u'hour08', u'hour09', u'hour10', u'hour11', u'hour12', u'hour13', u'hour14', u'hour15', u'hour16', u'hour17', u'hour18', u'hour19', u'hour20', u'hour21', u'hour22', u'hour23', u'hour24']]
                f['plan_date'] = f['plan_date'].str.split().apply(lambda x: x[0]).apply(lambda x: datetime.datetime.strptime(x, '%m/%d/%Y'))
                f = f.set_index('plan_date').stack().reset_index().rename(columns={'level_1':'hour', 0:'load'})
                f['hour'] = f['hour'].str.replace('hour','').astype(int)-1
                f['date'] = f.apply(lambda x: datetime.datetime(x['plan_date'].year, x['plan_date'].month, x['plan_date'].day, x['hour']), axis=1)
                f = pd.DataFrame(f.set_index('date')['load'])
                df = pd.concat([df, f], axis=0)
        
	df['year'] = [i.year for i in df.index]
        df_d.update({u : df})
```

```python
def plot_curvefit(xdata, ydata):
    xy = np.vstack([xdata,ydata])
    z = gaussian_kde(xy)(xy)
    scatter(xdata, ydata, c=z, s=100, edgecolor='')
```

```python
#### Regress by year for each utility

#### SRP

srp_regress = pd.concat([az_census['Maricopa'], df_d['srp'].groupby('year').mean()], axis=1, join='inner')

a, b = np.polyfit(srp_regress['Maricopa'], srp_regress['load'], 1)
y, p = np.polyfit(range(len(srp_regress)), srp_regress['Maricopa'], 1)
df_d['srp']['pop_est'] = [((y/(365*24))*x + p) for x in range(len(df_d['srp']))]
df_d['srp']['load_trend'] = df_d['srp']['pop_est'].apply(lambda x: a*x + b)
df_d['srp']['load_detrend'] = df_d['srp']['load'] - df_d['srp']['load_trend']

#### APS

aps_regress = pd.concat([az_census.sum(axis=1), df_d['aps'].groupby('year').mean()], axis=1, join='inner')

a2, b2 = np.polyfit(aps_regress[0], aps_regress['load'], 1)
y2, p2 = np.polyfit(range(len(aps_regress)), aps_regress[0], 1)
df_d['aps']['pop_est'] = [((y2/(365*24))*x + p2) for x in range(len(df_d['aps']))]
df_d['aps']['load_trend'] = df_d['aps']['pop_est'].apply(lambda x: a2*x + b2)
df_d['aps']['load_detrend'] = df_d['aps']['load'] - df_d['aps']['load_trend']

#### LDWP

ldwp_regress = pd.concat([ca_census['LosAngeles'], df_d['ldwp'].groupby('year').mean()], axis=1, join='inner')

a3, b3 = np.polyfit(ldwp_regress['LosAngeles'], ldwp_regress['load'], 1)
y3, p3 = np.polyfit(range(len(ldwp_regress)), ldwp_regress['LosAngeles'], 1)
df_d['ldwp']['pop_est'] = [((y3/(365*24))*x + p3) for x in range(len(df_d['ldwp']))]
df_d['ldwp']['load_trend'] = df_d['ldwp']['pop_est'].apply(lambda x: a3*x + b3)
df_d['ldwp']['load_detrend'] = df_d['ldwp']['load'] - df_d['ldwp']['load_trend']
```

```python
import string

basepath_861 = '/home/akagi/Documents/eia_form_861'

path_d = {
1990: 'f86190',
1991: 'f86191',
1992: 'f86192',
1993: 'f86193',
1994: 'f86194',
1995: 'f86195',
1996: 'f86196',
1997: 'f86197',
1998: 'f86198',
1999: 'f86199',
2000: 'f86100',
2001: '2001',
2002: '2002',
2003: '2003',
2004: '2004',
2005: '2005',
2006: 'f86106',
2007: '2007',
2008: '2008',
2009: '2009',
2010: 'f86110',
2011: 'f86111',
2012: 'f8612012',
}

#util = {
#'aps' : 803,
#'srp' : 16572,
#'ldwp' : 11208
#        }
#util_2006 = {
#'aps' : 116,
#'srp' : 244,
#'ldwp' : 194

#}

sales = pd.DataFrame()

#pre_1999 = pd.DataFrame()

def make_sales():
    
    pre_1999 = pd.DataFrame()
    post_1999 = pd.DataFrame()
    post_2001 = pd.DataFrame()
    post_2007 = pd.DataFrame()
    post_2008 = pd.DataFrame()
    
    for y in path_d.keys():
        if 1990 <= y <= 1991:
            df = pd.read_excel('%s/%s/F861TYP1.xls' % (basepath_861, path_d[y]))
            df['YEAR'] = y
            pre_1999 = pre_1999.append(df)
        if 1992 <= y <= 1998:
            df = pd.read_excel('%s/%s/F861TYP1.xls'  % (basepath_861, path_d[y]))
            df['YEAR'] = y
            pre_1999 = pre_1999.append(df)

    #### File_ format starts here
    
        if y == 1999:
            df = pd.read_excel('%s/%s/FILE2.xls' % (basepath_861, path_d[y]))
            df['YEAR'] = y
            post_1999 = post_1999.append(df)
        if y == 2000:
            df = pd.read_excel('%s/%s/FILE2.xls' % (basepath_861, path_d[y]))
            df['YEAR'] = y
            post_1999 = post_1999.append(df)

        if 2001 <= y <= 2006:
            df = pd.read_excel('%s/%s/file2.xls' % (basepath_861, path_d[y]))
            df['YEAR'] = y
            post_2001 = post_2001.append(df)

        if y == 2007:
            df = pd.read_excel('%s/%s/file2_%s.xls' % (basepath_861, path_d[y], y))
#            df['YEAR'] = y
            post_2007 = post_2007.append(df)            
        if 2008 <= y <= 2011:
            df = pd.read_excel('%s/%s/file2_%s.xls' % (basepath_861, path_d[y], y), skiprows=2)
            df['YEAR'] = y
            post_2008 = post_2008.append(df)
            
#### REFORMAT PRE-1999

    c_ser = pd.Series(pre_1999.columns)
    test = pre_1999.loc[:, c_ser[c_ser.str.contains('[A-Z]\d_|UTILCODE|YEAR')]]
    t1 = test.loc[:, pd.Series(test.columns)[pd.Series(test.columns).str.contains('1_|UTILCODE|YEAR')]]
    t2 = test.loc[:, pd.Series(test.columns)[pd.Series(test.columns).str.contains('2_|UTILCODE|YEAR')]]
    t1.columns = pd.Series(t1.columns).str.replace('1_', '_')
    t2.columns = pd.Series(t2.columns).str.replace('2_', '_')
    pre_1999_r = pd.concat([t1,t2]).rename(columns={'STCODE_1':'STATE'})#.groupby(['UTILCODE', 'STCODE_1']).mean()
    pre_1999_r.columns = pd.Series(pre_1999_r.columns).str.replace('_1', '_RES').str.replace('_2', '_COM').str.replace('_3', '_IND').str.replace('_4', '_HWY').str.replace('_5', '_OTH').str.replace('_6', '_TOT')

#### REFORMAT POST-1999

    post_1999.columns = pd.Series(post_1999.columns).str.replace('([A-Z]+)(SALES).*', '\\2_\\1').str.replace('([A-Z]+)(CONS).*', '\\2_\\1').str.replace('([A-Z]+)(REV).*', '\\2_\\1').str.replace('CONS_', 'CONSUM_').str.replace('SALES', 'MWH')

#### REFORMAT POST-2001   

    post_2001.columns = pd.Series(post_2001.columns).str.upper().str.replace('([A-Z]+) (REV).*', '\\2_\\1').str.replace('([A-Z]+) (SALES).*', '\\2_\\1').str.replace('([A-Z]+) (CONSUM).*', '\\2_\\1').str.replace('TRANS', 'HWY').str.replace('SALES', 'MWH').str.replace('UTILITY_ID', 'UTILCODE').str.replace('TOTAL', 'TOT')

#### REFORMAT POST-2007

    post_2007.columns = pd.Series(post_2007.columns).str.upper().str.replace('([A-Z]+)_(REV).*', '\\2_\\1').str.replace('([A-Z]+)_(SALES).*', '\\2_\\1').str.replace('([A-Z]+)_(CONSUM).*', '\\2_\\1').str.replace('RESIDENTIAL', 'RES').str.replace('COMMERCIAL', 'COM').str.replace('INDUSTRIAL', 'IND').str.replace('TRANSPORTATION', 'HWY').str.replace('UTILITY_ID', 'UTILCODE').str.replace('STATE_CODE', 'STATE').str.replace('SALES', 'MWH').str.replace('TOTAL', 'TOT')

#### REFORMAT POST-2008

    post_2008.columns = pd.Series(post_2008.columns).str.upper().str.replace('THOUSANDS DOLLARS', 'REV_').str.replace('MEGAWATTHOURS', 'MWH_').str.replace('COUNT', 'CONSUM_').str.replace('_$', '_RES').str.replace('\.1', 'COM').str.replace('\.2', 'IND').str.replace('\.3', 'HWY').str.replace('\.4', 'TOT').replace('UTILITY NUMBER', 'UTILCODE')
    
    df_return = pd.concat([pre_1999_r, post_1999, post_2001, post_2007, post_2008], join='inner')
    
    for i in df_return.iloc[:, 3:]:
        df_return[i] = df_return[i].astype(str).str.replace('\.\d*', '').str.replace('[^0-9]', '').replace('', np.nan).astype(float)
        
    return df_return
```
