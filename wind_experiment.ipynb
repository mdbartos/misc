```python
import pandas as pd
import numpy as np
import os
```

```python
nrel_path = '/home/tabris/Downloads/nrel_wind'

df_2004 = pd.read_csv('%s/%s/1114.csv' % (nrel_path, 2004), skiprows=1, names=['date', 'wind', 'mw', 'score', 'correct'])[['date', 'wind']]
df_2005 = pd.read_csv('%s/%s/1114.csv' % (nrel_path, 2005), skiprows=1, names=['date', 'wind', 'mw', 'score', 'correct'])[['date', 'wind']]
df_2006 = pd.read_csv('%s/%s/1114.csv' % (nrel_path, 2006), skiprows=1, names=['date', 'wind', 'mw', 'score', 'correct'])[['date', 'wind']]

df = pd.concat([df_2004, df_2005, df_2006])
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['month'] = [i.month for i in df['date']]
df['day'] = [i.day for i in df['date']]
df['hour'] = [i.hour for i in df['date']]
df['minute'] = [i.minute for i in df['date']]
```

```python
g = df.groupby(['month', 'day']).mean()['wind']
```

```python
%pylab inline
```

```python
plot(g)
```

```python
h = df.groupby(['month', 'day', 'hour', 'minute']).mean()['wind']
hmax = df.groupby(['month', 'day', 'hour', 'minute']).max()['wind'].values
hmin = df.groupby(['month', 'day', 'hour', 'minute']).min()['wind'].values
```

```python
#fill_between(range(len(hmax)), hmax, hmin, facecolor='0.85')
plot(hmax, alpha=0.85)
plot(hmin, alpha=0.85)
plot(h, alpha=0.85)
```

```python
#fill_between(range(len(hmax)), hmax, hmin, facecolor='0.85')
plot(hmax, alpha=0.85)
plot(hmin, alpha=0.85)
plot(h, alpha=0.85)
```
