## Function definitions

```python
def return_cent(shpfile):
    s = pd.Series(range(len(shpfile)))
    def poly_cent(n):
        return geometry.Polygon(shpfile[n]['geometry']['coordinates'][0]).centroid.coords[:][0]
    def mpoly_cent(n):
        multi_cent = pd.Series(range(len(shpfile[n]['geometry']['coordinates'])))
        a = multi_cent.apply(lambda j: geometry.Polygon(shpfile[n]['geometry']['coordinates'][j][0]).area)
        lons = multi_cent.apply(lambda j: geometry.Polygon(shpfile[n]['geometry']['coordinates'][j][0]).centroid.coords[:][0][0])
        lats = multi_cent.apply(lambda j: geometry.Polygon(shpfile[n]['geometry']['coordinates'][j][0]).centroid.coords[:][0][1])
        x_cent = (a*lons).sum()/a.sum()
        y_cent = (a*lats).sum()/a.sum()
        return tuple([x_cent, y_cent])
    return (s.apply(lambda x: poly_cent(x) if shpfile[x]['geometry']['type'] == 'Polygon' else mpoly_cent(x) if shpfile[x]['geometry']['type'] == 'MultiPolygon' else None)).apply(pd.Series).rename(columns={0:'x', 1:'y'})
```

```python
def return_bounds(shpfile):
    s = pd.Series(range(len(shpfile)))
    def poly_bounds(n):
        return geometry.Polygon(shpfile[n]['geometry']['coordinates'][0]).bounds
    def mpoly_bounds(n):
        multi_s = pd.Series(range(len(shpfile[n]['geometry']['coordinates'])))
        multi_bounds = multi_s.apply(lambda j: geometry.Polygon(shpfile[n]['geometry']['coordinates'][j][0]).bounds).apply(pd.Series)
        return tuple(multi_bounds[[0,1]].min().append(multi_bounds[[2,3]].max()))

    return (s.apply(lambda x: poly_bounds(x) if shpfile[x]['geometry']['type'] == 'Polygon' else mpoly_bounds(x) if shpfile[x]['geometry']['type'] == 'MultiPolygon' else None)).apply(pd.Series).rename(columns={0:'xmin', 1:'ymin', 2:'xmax', 3:'ymax'})
```

```python
def geom_types(shp):
    s = pd.Series(range(len(shp)))
    return s.apply(lambda x: shp[x]['geometry']['type'])
```

```python
def pt_in_bounds(x_coord, y_coord, shp_bounds):
    return list(shp_bounds.loc[(shp_bounds['xmin'] < x_coord) & (shp_bounds['ymin'] < y_coord) & (shp_bounds['xmax'] > x_coord) & (shp_bounds['ymax'] > y_coord)].index)
```

## Main function

```python
def centroid_in_poly(shp1, shp2):
    shp1_c = return_cent(shp1)
    shp2_bounds = return_bounds(shp2)
    c_memb = shp1_c.apply(lambda x: pt_in_bounds(x['x'], x['y'], shp2_bounds), axis=1).apply(pd.Series)
    c_memb.columns = [str(i) for i in c_memb.columns]

    def points_in_poly():
        geom = geom_types(shp2)
        ct_c = shp1_c.apply(geometry.Point, axis=1)
        
        c_poly = c_memb.where(c_memb.apply(lambda x: x.map(geom) == 'Polygon')).dropna(how='all').dropna(axis=1, how='all').stack().astype(int).reset_index().set_index(0).sort_index()['level_0'].reset_index()        

        bool_df = pd.Series()

        for i in pd.Series(c_poly[0]).astype(int).unique():
            m = geometry.Polygon(shp2[i]['geometry']['coordinates'][0])
            bool_df = pd.concat([bool_df, c_poly.loc[c_poly[0] == i, 'level_0'].apply(lambda x: m.contains(ct_c.loc[x]))])
            
        return c_poly[bool_df].set_index('level_0')
    
    def mpoly_cu(idxno):
        polylist = []
        for i in range(len(shp2[idxno]['geometry']['coordinates'])):
            polylist.append(geometry.Polygon(shp2[idxno]['geometry']['coordinates'][i][0]))
            
        return ops.cascaded_union(polylist)
    
    def points_in_mpoly():
        geom = geom_types(shp2)
        ct_c = shp1_c.apply(geometry.Point, axis=1)
        c_mpoly = c_memb.where(c_memb.apply(lambda x: x.map(geom) == 'MultiPolygon')).dropna(how='all').dropna(axis=1, how='all').stack().astype(int).reset_index().set_index(0).sort_index()['level_0'].reset_index()        

        bool_df = pd.Series()

        for i in pd.Series(c_mpoly[0]).astype(int).unique():
            m = mpoly_cu(i)
            bool_df = pd.concat([bool_df, c_mpoly.loc[c_mpoly[0] == i, 'level_0'].apply(lambda x: m.contains(ct_c.loc[x]))])
            
        return c_mpoly[bool_df].set_index('level_0')
    
    return_df = pd.concat([pd.DataFrame(points_in_poly()), points_in_mpoly()]).sort_index().reindex(range(len(shp1)))
    return_df.index.rename('shape1', inplace=True)
    return_df.rename(columns={0:'shape2'}, inplace=True)
    
    return return_df
```

### Get centroids using "for" loop: slower

```python
%%timeit

df = pd.DataFrame(columns=['lon', 'lat'])

for i in range(len(states)):
    if states[i]['geometry']['type'] == 'Polygon':
        df.loc[i] = geometry.Polygon(states[i]['geometry']['coordinates'][0]).centroid.coords[:][0]
    elif states[i]['geometry']['type'] == 'MultiPolygon':
        multi_cent = pd.DataFrame(columns=['a', 'lon', 'lat'])
        for j in range(len(states[i]['geometry']['coordinates'])):
            lon = geometry.Polygon(states[i]['geometry']['coordinates'][j][0]).centroid.coords[:][0][0]
            lat = geometry.Polygon(states[i]['geometry']['coordinates'][j][0]).centroid.coords[:][0][1]
            multi_cent.loc[j, 'a'] = geometry.Polygon(states[i]['geometry']['coordinates'][j][0]).area
            multi_cent.loc[j, ['lon', 'lat']] = lon, lat
        x_cent = (multi_cent['a']*multi_cent['lon']).sum()/multi_cent['a'].sum()
        y_cent = (multi_cent['a']*multi_cent['lat']).sum()/multi_cent['a'].sum()
        df.loc[i] = x_cent, y_cent
```
