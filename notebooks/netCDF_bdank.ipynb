### Import modules

```python
import numpy as np
import pandas as pd
import netCDF4
import os
import datetime
```

### Specify home directory

```python
homedir = os.path.expanduser('~')
```

### Load NetCDF4 dataset

Maximum daily temperature, Maricopa County area, 2030-2050

```python
tmax5 = netCDF4.Dataset('%s/tmax_5.nc' % homedir)
```

### Get integer indices of projection scenarios

```python
proj = pd.Series(tmax5.Projections.split(',')).str.strip()[:-1]

proj
```

### Let's use projection scenario 116: mpi-esm-lr.2.rcp45

```python
tmax = np.ma.filled(tmax5.variables['tasmax'][116,:,:,:], np.nan)
```

### Convert 3D array to a pandas panel

```python
tmax = pd.Panel(tmax,
                items=pd.date_range(datetime.date(2030,1,1), freq='d', periods=tmax5.variables['tasmax'].shape[1]),
                major_axis=tmax5.variables['latitude'][:],
                minor_axis=tmax5.variables['longitude'][:] - 360)
```

### Select summertime entries

```python
tmax_summer = tmax[np.in1d(tmax.items.month, [6,7,8])]
```

### Get mean summertime temperature

```python
tmax_mean = tmax_summer.mean(axis=0)

tmax_mean
```

### Get 90th percentile of summertime temperature

```python
tmax_90 = tmax_summer.to_frame().quantile(0.9, axis=1).unstack()

tmax_90
```
