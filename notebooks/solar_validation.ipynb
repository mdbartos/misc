```python
import pandas as pd
import numpy as np
import datetime
```

```python
mkdate = lambda x: datetime.date(int(x['YEAR']), int(x['MONTH']), int(x['DAY']))
```

```python
hist = pd.read_csv('/home/chesterlab/Bartos/post/pv/hist.529', sep='\t')
ukmo_a1b = pd.read_csv('/home/chesterlab/Bartos/post/pv/ukmo_a1b.529', sep='\t')
ukmo_a2 = pd.read_csv('/home/chesterlab/Bartos/post/pv/ukmo_a2.529', sep='\t')
ukmo_b1 = pd.read_csv('/home/chesterlab/Bartos/post/pv/ukmo_b1.529', sep='\t')
```

```python
hist['date'] = hist.apply(mkdate, axis=1)
hist = hist.set_index('date')
ukmo_a1b['date'] = ukmo_a1b.apply(mkdate, axis=1)
ukmo_a1b = ukmo_a1b.set_index('date')
ukmo_a2['date'] = ukmo_a2.apply(mkdate, axis=1)
ukmo_a2 = ukmo_a2.set_index('date')
ukmo_b1['date'] = ukmo_b1.apply(mkdate, axis=1)
ukmo_b1 = ukmo_b1.set_index('date')
```

```python
hist['OUT_R_NET'].describe()
```

```python
ukmo_a1b['OUT_R_NET'].describe()
```

```python
plot(hist.index, hist['OUT_R_NET'], color='blue')
plot(ukmo_a1b.index, ukmo_a1b['OUT_R_NET'], color='red')
```

```python
plot(hist.groupby('MONTH').mean().index, hist.groupby('MONTH').mean()['OUT_R_NET'])
plot(ukmo_a1b.groupby('MONTH').mean().index, ukmo_a1b.groupby('MONTH').mean()['OUT_R_NET'])
plot(ukmo_a2.groupby('MONTH').mean().index, ukmo_a2.groupby('MONTH').mean()['OUT_R_NET'])
plot(ukmo_b1.groupby('MONTH').mean().index, ukmo_b1.groupby('MONTH').mean()['OUT_R_NET'])
xlim(1,12)
```

```python
histsummer = hist.loc[hist['MONTH'].isin([6,7,8])]
a1bsummer = ukmo_a1b.loc[ukmo_a1b['MONTH'].isin([6,7,8])]
a2summer = ukmo_a2.loc[ukmo_a2['MONTH'].isin([6,7,8])]
b1summer = ukmo_b1.loc[ukmo_a2['MONTH'].isin([6,7,8])]
plot(histsummer.index, histsummer['OUT_R_NET'], color='blue')
plot(a1bsummer.index, a1bsummer['OUT_R_NET'], color='red')
```
