```python
import fiona
import pandas as pd
import numpy as np
```

# Read shapefile

```python
states = fiona.open('/home/tabris/GIS/states.shp', 'r')
```

# Plot it

```python
%pylab inline
```

```python
for i in range(len(states[1]['geometry']['coordinates'])):
    rec = np.asarray(pd.DataFrame.from_dict(states[1]['geometry']).loc[i, 'coordinates'])
    rec = rec.reshape(rec.shape[1], rec.shape[2])
    plot(rec[:,0], rec[:,1])
```

```python
for s in range(len(states)):
    for i in range(len(states[s]['geometry']['coordinates'])):
        rec = np.asarray(pd.DataFrame.from_dict(states[s]['geometry']).loc[i, 'coordinates'])
        if len(rec.shape) > 2:
            rec = rec.reshape(rec.shape[1], rec.shape[2])
        plot(rec[:,0], rec[:,1])
```
