```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
C_0 = 6
V_0 = 400
v = 1
x = 8
n = 30
```

```python
c_k = C_0
v_k = V_0
C = []
M = []
V = []

for k in range(n):
    m_k = (v_k*c_k - x*k*v*c_k)
    v_k =(V_0 - k*v)
    if v_k < 1e-4:
        break
    c_k = m_k / v_k
    C.append(c_k)
    M.append(m_k)
    V.append(v_k)

C = np.asarray(C)
M = np.asarray(M)
V = np.asarray(V)
```

```python
plt.plot(np.arange(len(C)), C)
plt.title('Concentration vs. drips')
plt.ylabel('Concentration')
plt.xlabel('Drips (k)')
```

```python
plt.plot(np.arange(len(C)), V)
plt.title('Volume vs. drips')
plt.ylabel('Volume')
plt.xlabel('Drips (k)')
```

```python
plt.plot(np.arange(len(C)), M)
plt.title('Mass of solute vs. drips')
plt.ylabel('Mass of solute')
plt.xlabel('Drips (k)')
```

```python
plt.plot(np.cumsum(C) / np.arange(1, len(C)+1))
plt.title('Concentration in collector')
plt.ylabel('Concentration in collector')
plt.xlabel('Drips (k)')
```
