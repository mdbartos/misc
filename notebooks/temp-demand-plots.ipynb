```python
def plot_kd(xdata, ydata):
    xy = np.vstack([xdata,ydata])
    z = scipy.stats.gaussian_kde(xy)(xy)
    plt.scatter(xdata, ydata, c=z, s=100, edgecolor='')
```

### Weighted Average

```python
WECC_customers = customers.loc[d.minor_axis.values]['CUSTOMERS']
tot_WECC_cust = WECC_customers.sum()
```

```python
WECC_customers
```

```python
annual_scenarios = d.to_frame().unstack().loc['1990':].resample('A')
```

```python
annual_trend = annual_scenarios.div(annual_scenarios.loc[:'2010'].mean()).stack(level=0)
```

```python
annual_trend = (annual_trend.multiply(WECC_customers, axis=1).sum(axis=1)/tot_WECC_cust).unstack()
```

```python
annual_trend.plot(legend=False)
axvline(datetime.date(2010,1,1))
```

```python
scenario_means = pd.concat([annual_trend.loc[:, rcp26_labels].mean(axis=1), annual_trend.loc[:, rcp45_labels].mean(axis=1), annual_trend.loc[:, rcp85_labels].mean(axis=1)], axis=1)
scenario_means.columns = ['rcp26', 'rcp45', 'rcp85']
annual_max = annual_trend.resample('A').max(axis=1)
annual_min = annual_trend.resample('A').min(axis=1)

annual_max = pd.rolling_mean(annual_max, 5).loc['2010':]
annual_min = pd.rolling_mean(annual_min, 5).loc['2010':]

pd.rolling_mean(scenario_means.resample('A'), 5).loc['2010':].plot(color=['blue', 'orange', 'red'], legend=False)
scenario_means.resample('A').loc[:'2010', 'rcp26'].plot(color='black', label='hist')
fill_between(annual_max.index, annual_max, annual_min, color='0.85')

xlabel('Year')
ylabel('Normalized Per Capita Load')
title('Normalized Per Capita Load')
axvline(datetime.date(2010,1,1), color='black', linestyle='--')
legend(loc=2)
```
